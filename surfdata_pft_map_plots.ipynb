```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import cartopy.crs as ccrs
```

```python
fin = xr.open_dataset('surfdata_0.9x1.25_16pfts_Irrig_CMIP6_simyr2000_c170824.nc')
```

```python
fin = fin.rename_dims({'lsmlon':'lon', 'lsmlat':'lat'})
fin['lat'] = fin['LATIXY'][:,0]
fin['lon'] = fin['LONGXY'][0,:]
```

```python
fg = fin['PCT_NAT_PFT'].isel(natpft=slice(0,15)).plot(x='lon',y='lat',col='natpft', col_wrap=3, levels=[0.,1.,2.,5.,10.,20.,50.], cmap='Greens', transform=ccrs.PlateCarree(), subplot_kws=dict(projection=ccrs.Robinson()), add_colorbar=False)

for ax in fg.axes.flat:
    ax.coastlines() 
    
fg.add_colorbar(orientation="vertical", shrink=0.6, anchor=(1.5, 0.5))

plt.subplots_adjust(hspace=-0.65)
```
